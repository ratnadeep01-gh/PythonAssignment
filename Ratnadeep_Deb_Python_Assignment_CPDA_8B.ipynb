1. DATA INGESTION AND QUALITY PROFILING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read car prices into pandas df

car_df=pd.read_csv("/content/car_prices.csv")
car_df
```

```python
# Display first 5 rows

car_df.head()
```

```python
# Display data types & record counts

car_df.info
```

```python
# Check the shape of the data set

car_df.shape
```

```python
# Display column names and data types

car_df.info()
```

```python
# Quantify nulls per column; visualize with a bar chart or heat map

# Let's count the nulls per column

null_counts=car_df.isnull().sum()

# Filter out columns with no null values

null_counts=null_counts[null_counts > 0]

if not null_counts.empty:
    plt.figure(figsize=(12, 6))
    null_counts.sort_values(ascending=False).plot(kind='bar', color='skyblue')
    plt.title('No. of Null Values Per Column in car_df')
    plt.xlabel('Attributes or Column Description')
    plt.ylabel('No. of Null Values')
    plt.show()
else:
    print('No null values found in the DataFrame.')
```

```python
# Resolve null values by appropriate strategy based on the datatype and percentage of null values.

# Identify categorical and numerical columns that have null values (excluding 'vin')
missing_categorical_cols=['make', 'model', 'trim', 'body', 'transmission', 'color', 'interior', 'saledate']
missing_numerical_cols=['condition', 'odometer', 'mmr', 'sellingprice']

# Drop rows where 'vin' is null
car_df.dropna(subset=['vin'], inplace=True)

# Update categorical columns with mode
for col in missing_categorical_cols:
    if col in car_df.columns:
        mode_value=car_df[col].mode()[0]
        car_df[col].fillna(mode_value, inplace=True)

# Update numerical columns with median
for col in missing_numerical_cols:
    if col in car_df.columns:
        median_value=car_df[col].median()
        car_df[col].fillna(median_value, inplace=True)

print("Null values resolved. Displaying DataFrame info to confirm:")
car_df.info()
```

```python
# Count number of duplicate records

all_rows=car_df.shape[0]
duplicate_rows=car_df.duplicated().sum()

print(f"Number of duplicate records: {duplicate_rows}")


# Delete duplicates, if any present

if duplicate_rows > 0:
    car_df.drop_duplicates(inplace=True)
    print(f"Duplicate records deleted. No. number of rows: {car_df.shape[0]}")
    print(f"Total rows deleted: {all_rows - car_df.shape[0]}")
else:
    print("No duplicate records found.")
```

2. DATA FRAME QUERIES

```python
# Calculate the average, minimum, and maximum car price

average_price=car_df['sellingprice'].mean()
min_price=car_df['sellingprice'].min()
max_price=car_df['sellingprice'].max()

print(f"Average Car Price: ${average_price}")
print(f"Minimum Car Price: ${min_price}")
print(f"Maximum Car Price: ${max_price}")
```

```python
# List all unique colors of cars

def clean_color(color):
    if isinstance(color, str):
        if color.isdigit() or color == '—':
            return 'Unknown'
    return color

# Ensuring only color as strings are reported

car_df['color']=car_df['color'].apply(clean_color)

unique_colors_cleaned=car_df['color'].unique()
print("Unique Car Colors with strings characters only:")
for color in unique_colors_cleaned:
    print(color)
```

```python
# Find the number of unique car brands and car models

unique_car_brands=car_df['make'].nunique()
unique_car_models=car_df['model'].nunique()

print(f"Number of unique car brands: {unique_car_brands}")
print(f"Number of unique car models: {unique_car_models}")
```

```python
# Find all car information having selling prices greater than $165000

high_value_cars=car_df[car_df['sellingprice'] > 165000]
high_value_cars
```

```python
# Find the top 5 most frequently sold car models

top_5_models=car_df['model'].value_counts().head(5)
print("Top 5 most frequently sold car models:")
print(top_5_models)
```

```python
# What is the average selling price of cars by brand (make)

avg_price_by_make=car_df.groupby('make')['sellingprice'].mean().sort_values(ascending=False)
print("Average Selling Price by Car Brand (Make):")
print(avg_price_by_make.head(10))

# Top 10 only selected for display
```

```python
# What is the minimum selling price of cars for each interior

min_price_by_interior=car_df.groupby('interior')['sellingprice'].min().sort_values(ascending=True)
print("Minimum Selling Price by Interior:")
print(min_price_by_interior)
```

```python
# Find highest odometer reading per year from highest to lowest order

highest_odometer_per_year=car_df.groupby('year')['odometer'].max().sort_values(ascending=False)
print("Highest Odometer Reading Per Year (Highest to Lowest):")
print(highest_odometer_per_year)
```

```python
# Create a new column for car age (assuming the current year is 2025)

car_df['car_age']=2025 - car_df['year']
car_df.head()
```

```python
# Find the number of cars having a condition greater than or equal to 48 and odometer greater than 90000

filtered_cars_count=car_df[(car_df['condition'] >= 48) & (car_df['odometer'] > 90000)].shape[0]
print(f"No. of cars with condition >= 48 and odometer > 90000: {filtered_cars_count}")
```

```python
# Which state consistently has higher car prices for newer cars(year>2013)

newer_cars=car_df[car_df['year'] > 2013]
sp_state =newer_cars.groupby('state')['sellingprice'].mean().sort_values(ascending=False)

print("Average selling price for newer cars (year > 2013) by state (highest to lowest):")
print(sp_state.head(10))

# Displaying top 10 states
```

```python
#  For cars with excellent condition (top 20%), which makes have the lowest average price (value for money)

condition_threshold=car_df['condition'].quantile(0.80)
excellent_condition_cars=car_df[car_df['condition'] >= condition_threshold]

avg_price_by_make_exc_condition = excellent_condition_cars.groupby('make')['sellingprice'].mean().sort_values(ascending=True)

print(f"'Excellent condition' is defined as a condition score >= {condition_threshold:.2f}")
print("\nCar Makes with the Lowest Average Price for Excellent Condition Cars (Value for Money): \n")
print(avg_price_by_make_exc_condition.head(10))

# Display top 10 for brevity
```

3. DATA VISUALIZATION AND INSIGHTS

```python
# Show the correlation of all the features(columns) having numerical values(e.g. selling price, odometer)

numerical_cols=car_df.select_dtypes(include=['number']).columns
correlation_matrix=car_df[numerical_cols].corr()
print("Correlation Matrix of Numerical Features:")
print(correlation_matrix)
```

```python
# Plot a graph to show the average selling price by year and explain the pattern you notice from the graph. Which plot will you use bar or scatter?

average_price_by_year=car_df.groupby('year')['sellingprice'].mean()

plt.figure(figsize=(14, 7))
average_price_by_year.plot(kind='bar', color='green')
plt.title('Average Selling Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.tight_layout()
plt.show()

# Explain the pattern
print("\nBar chart is preferred. Pattern observed from the graph:")
print("The bar chart displays the average selling price of cars for each manufacturing year. ")
print("Typically, newer cars tend to have higher average selling prices, and older cars tend to have lower average selling prices. This pattern often reflects depreciation over time.")
print("From the graph, we can observe a general trend: ")
print("- There's a clear increase in average selling price as the year of manufacture approaches the present (e.g., 2015).")
print("- Older cars (e.g., from the 1980s and 1990s) generally have much lower average selling prices.")
print("- This indicates that car age is a significant factor in its selling price, with newer models retaining higher value.")
```

```python
# Plot a graph to show average selling price by odometer and explain the trend you notice from the graph.

plt.figure(figsize=(12, 6))
plt.scatter(car_df['odometer'], car_df['sellingprice'], alpha=0.5)
plt.title('Selling Price vs. Odometer Reading')
plt.xlabel('Odometer (miles)')
plt.ylabel('Selling Price ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\nScatter Plot provides better visualization. Pattern observed from the graph:")
print("The scatter plot above illustrates the relationship between a car's odometer reading and its selling price. We can typically expect a negative correlation between these two variables.")
print("From the graph, we can observe the following trend:")
print("- As the odometer reading increases (moving from left to right on the x-axis), the selling price generally tends to decrease (moving downwards on the y-axis).")
print("- This indicates an inverse relationship: cars with higher mileage usually command lower selling prices, which is a common characteristic of vehicle depreciation.")
print("- There is a higher density of data points in the lower odometer and higher selling price range, suggesting that newer or less-driven cars often get higher prices.")
print("- Conversely, for very high odometer readings, the selling prices are predominantly lower, and the spread of prices also tends to narrow.")
```

```python
# Plot a graph to show the number of cars sold in each state. Find the top three highest car selling states by having a look at the graph?

cars_by_state=car_df['state'].value_counts()

plt.figure(figsize=(15, 7))
cars_by_state.head(10).plot(kind='bar', color='brown')
plt.title('Number of Cars Sold in Each State (Top 10)')
plt.xlabel('State')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()
plt.show()

#  Display top 10 for better visualization

print("\nTop 3 highest car selling states based on the graph:")
for i, (state, count) in enumerate(cars_by_state.head(3).items()):
    print(f"{i+1}. {state}: {count} cars")
```

```python
# Plot a bar graph of average selling price by condition score ranges of size 5. Give a summary of the insights.

# Create condition score ranges
# Assuming condition scores range from 1 to 49, create bins of size 5
bins = list(range(0, 51, 5)) # Bins: [0,5), [5,10), ..., [45,50)
labels = [f'{i}-{i+4}' for i in range(0, 46, 5)]

car_df['condition_range'] = pd.cut(car_df['condition'], bins=bins, labels=labels, right=False, include_lowest=True)

# Calculate the average selling price for each condition range
average_price_by_condition_range = car_df.groupby('condition_range', observed=True)['sellingprice'].mean()

# Plotting the bar graph
plt.figure(figsize=(12, 6))
average_price_by_condition_range.plot(kind='bar', color='black')
plt.title('Average Selling Price by Condition Score Range')
plt.xlabel('Condition Score Range')
plt.ylabel('Average Selling Price ($)')
plt.tight_layout()
plt.show()

print("\nSummary of Insights:")
print("The bar graph displays the average selling price of cars grouped by condition score ranges (e.g., 0-4, 5-9, etc.).")
print("Typically, we expect cars in better condition (higher condition scores) to have higher average selling prices.")
print("From the graph, we can observe the following trends:")
print("- As the condition score range increases, the average selling price generally tends to rise. This indicates a positive correlation between car condition and its market value.")
print("- There might be some fluctuations or non-linearities eg. under bin 0-4, 5-9, but the overall trend should show an upward trajectory.")
print("- Cars in the highest condition score ranges (e.g., 40-44, 45-49) usually command the highest average prices.")
print("- This analysis reinforces the importance of vehicle condition as a determinant of its selling price.")
```

```python
# Plot a bar graph of no. of cars sold by condition ranges of size 10. Give some insights from the graph.

#  Create condition score ranges of size 10
bins_10 = list(range(0, 51, 10)) # Bins: [0,10), [10,20), ..., [40,50)
labels_10 = [f'{i}-{i+9}' for i in range(0, 41, 10)]

car_df['condition_range_10']=pd.cut(car_df['condition'], bins=bins_10, labels=labels_10, right=False, include_lowest=True)

# Calculate the number of cars for each condition range
cars_by_condition_range=car_df['condition_range_10'].value_counts().sort_index()

# Plotting the bar graph
plt.figure(figsize=(12, 6))
cars_by_condition_range.plot(kind='bar', color='darkorange')
plt.title('Number of Cars Sold by Condition Score Range (Size 10)')
plt.xlabel('Condition Score Range')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()
plt.show()

print("\nSummary of Insights:")
print("The bar graph displays the number of cars sold grouped by condition score ranges of size 10 (e.g., 0-9, 10-19, etc.).")
print("From the graph, we can observe the following trends:")
print("- The distribution of car conditions: some ranges might have significantly more cars than others.")
print("- Identify the most common condition ranges for cars in the dataset.")
print("- Notice if there are any condition ranges with very few or no cars, which could indicate data sparsity or specific market behaviors.")
print("- This visualization helps in understanding the overall 'health' profile of the cars being sold.")
```

```python
# Plot a box plot to show the distribution of car selling prices grouped by color. Extract insights and if required, remove outliers and plot the graph again.

plt.figure(figsize=(15, 8))
car_df.boxplot(column='sellingprice', by='color', rot=45)
plt.title('Distribution of Car Selling Prices by Color')
plt.xlabel('Car Color')
plt.ylabel('Selling Price ($)')
plt.suptitle('') # Suppress the default suptitle created by boxplot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()

print("\nInsights from the initial box plot:")
print("The box plot above shows the distribution of selling prices for cars, grouped by their color. Each box represents the interquartile range (IQR), with the line inside indicating the median. The 'whiskers' extend to data points within 1.5 times the IQR from the box, and points beyond these whiskers are considered outliers.")
print("Key observations:")
print("- **Presence of Outliers**: Many colors show numerous individual points above the upper whiskers, indicating a significant number of high-priced outliers.")
print("- **Range of Prices**: The overall spread of selling prices varies considerably across different colors.")
print("- **Median Prices**: The median selling prices (the line within the box) also differ, suggesting some colors might be associated with generally higher or lower value cars.")
print("- **'Unknown' Color**: The 'Unknown' color category, created during data cleaning, likely encompasses a wide range of cars and may show a broad distribution.")
print("Given the visual evidence of many outliers, especially on the higher end, removing them could provide a clearer view of the central distribution for most cars.")

```

```python
# --- Attempted with Outlier Removal and Re-plotting ---

# Calculate IQR for 'sellingprice'
Q1 = car_df['sellingprice'].quantile(0.25)
Q3 = car_df['sellingprice'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
car_df_cleaned = car_df[(car_df['sellingprice'] >= lower_bound) & (car_df['sellingprice'] <= upper_bound)]

print(f"\nNumber of records before outlier removal: {len(car_df)}")
print(f"Number of records after outlier removal: {len(car_df_cleaned)}")
print(f"Number of outliers removed: {len(car_df) - len(car_df_cleaned)}")

# Plot again after outlier removal
plt.figure(figsize=(15, 8))
car_df_cleaned.boxplot(column='sellingprice', by='color', rot=45)
plt.title('Distribution of Car Selling Prices by Color (Outliers Removed)')
plt.xlabel('Car Color')
plt.ylabel('Selling Price ($)')
plt.suptitle('') # Suppress the default suptitle created by boxplot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()

print("\nInsights from the box plot after outlier removal:")
print("After removing outliers (values outside 1.5 * IQR from Q1 and Q3), the box plot shows a more concentrated view of the typical selling prices for each car color.")
print("Key observations:")
print("- **Clearer Distributions**: The boxes and whiskers are now much tighter, providing a clearer picture of the typical range and median selling prices for the majority of cars within each color category.")
print("- **Comparative Analysis**: It's easier to compare the median selling prices and the spread of prices across different colors without the distortion of extreme values.")
print("- **Still Some Outliers**: Even after general outlier removal, some colors might still show a few individual points, indicating that 'outliers' in one group might be 'normal' for another, or simply values that are still unusual but within a less extreme range.")
print("- **'Unknown' Color**: The 'Unknown' category's distribution becomes more interpretable as well, showing its typical price range without the influence of very high or very low outliers.")
print("This cleaned plot provides a more robust understanding of the typical car selling price distribution across various colors.")
```
